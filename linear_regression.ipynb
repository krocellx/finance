```python
import xlwings as xw
import statsmodels.api as sm
import pandas as pd
```

```python
book = xw.Book(r'F:\OneDrive\Rotman_MBA\Term_04\RSM2209_Financial_Statement_Analysis\Group_Assignment\Regression.xlsx')
sheet = book.sheets('Data')
```

```python
test_sht = book.sheets('Sheet4')
test_sht.range('A:B').clear()
```

```python
results_sht = book.sheets('TSLA_Regression_Result')
df_data = sheet.range('B6').options(pd.DataFrame, header=1, index=False, expand='table').value
df_data
```

```python
results_sht = book.sheets('F_Regression_Result')
df_data = sheet.range('K6').options(pd.DataFrame, header=1, index=False, expand='table').value
df_data
```

```python
results_sht = book.sheets('F_excl_2021Q4_Regression_Result')
df_data = sheet.range('T6').options(pd.DataFrame, header=1, index=False, expand='table').value
df_data
```

```python
results_sht.range('A:G').clear()
```

```python
# CFO_t = a + b * NI_t

Y = df_data[~df_data['CFO_t'].isna()]['CFO_t']
X = df_data[~df_data['CFO_t'].isna()]['NI']
X = sm.add_constant(X) # adding a constant

model = sm.OLS(Y, X).fit()
predictions = model.predict(X) 

print_model = model.summary()

last_row = results_sht.range('A' + str(results_sht.cells.last_cell.row)).end('up').row

results_sht.range('A'+str(last_row + 3)).value = 'Model 1: CFO_t = a + b * NI_t'

results_as_html_01 = print_model.tables[0].as_html()
results_sht.range('A'+str(last_row + 5)).options(index=False).value = pd.read_html(results_as_html_01, header=0)[0]

results_as_html_02 = print_model.tables[2].as_html()
results_sht.range('A'+str(last_row + 15)).options(index=False).value = pd.read_html(results_as_html_02, header=0)[0]

results_as_html_03 = print_model.tables[1].as_html()
results_sht.range('A'+str(last_row + 20)).options(index=False).value = pd.read_html(results_as_html_03, header=0)[0]

print(print_model)
```
